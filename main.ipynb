```python
from transformers import AutoTokenizer, AutoModelForCausalLM
import torch

print(torch.cuda.is_available())
ACCESS_TOKEN = "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"
PROMPT_TEMPLATE= "{input} \n Answer: "
```

```python

model_name = "meta-llama/Llama-3.1-8B"
tokenizer = AutoTokenizer.from_pretrained(model_name, token = ACCESS_TOKEN)
model = AutoModelForCausalLM.from_pretrained(model_name, token = ACCESS_TOKEN)
```

```python
#user_input = input("Enter your question: ")
user_input = "what is 1 plus 1, answer concisely?"
prompt = PROMPT_TEMPLATE.format(input = user_input)
print(model.generation_config)
output_id = ""
curr_sequence = tokenizer(prompt, return_tensors = "pt").input_ids

while True:

  #return (batch, sequence, vocab score)
  raw_output = model(curr_sequence)
  output_id = torch.argmax(raw_output.logits[:, -1, :], dim = -1)
  print(output_id.item())
  curr_sequence = torch.cat([curr_sequence ,output_id.unsqueeze(0)], dim = -1)
  print(tokenizer.decode(curr_sequence.flatten()))
  if output_id.item() == tokenizer.eos_token_id:
    break
```
